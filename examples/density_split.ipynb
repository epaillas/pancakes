```python
import pancakes.densitysplit.box as ds
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# generate random centres for split densities
ncentres = 1_000_000
sampling = 'uniform'
xmin, xmax = 0, 1500
ymin, ymax = 0, 1500
zmin, zmax = 0, 1500
output_format = 'unformatted'
output_filename =  'data/ds_centres/ds_centres.unf'

centres = ds.generate_centres(ncentres=ncentres, sampling=sampling, xmin=xmin,
                   xmax=xmax, ymin=ymin, ymax=ymax, zmin=zmin, zmax=zmax,
                   output_format=output_format, output_filename=output_filename)

fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(centres[::100,0], centres[::100,1], s=1.0)
ax.set_xlabel('x [Mpc/h]')
ax.set_ylabel('y [Mpc/h]')
ax.set_title('Random points for split densities')
plt.show()
```
